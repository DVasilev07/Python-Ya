```python
import pandas as pd
import numpy as np
import sklearn 
from sklearn import feature_selection
import matplotlib.pyplot as plt
from sklearn.cluster.hierarchical import AgglomerativeClustering
```

### Работа с сотами


```python
must_columns_avg_for_test = ['T_DATE','CELL_LAC_ID', 'RAB_CS_DROP_RATE_3G']
must_columns_avg = ['T_DATE','CELL_LAC_ID','RAB_CS_DROP_RATE_3G','RAB_PS_DROP_RATE_3G','TBF_DROP_RATE_2G','TCH_DROP_RATE_2G']
```

```python
data_sota_avg_test_all = pd.read_csv('bs_avg_kpi.csv',';',decimal=',',parse_dates=['T_DATE'],nrows = 100,usecols = must_columns_avg)
```

```python
# Работа с форматом day.month
ar = np.array(data_sota_avg_test_all['T_DATE'])
month_ar = []
day_ar = []
for i in range(len(ar)):
    month = ar[i][3:5]
    day = ar[i][0:2]
    month_ar.append(month)
    day_ar.append(day)
    
data_sota_avg_test_all['month'] = month_ar
data_sota_avg_test_all['day'] = day_ar
data_sota_avg_test_all.sort_values(['CELL_LAC_ID','month','day'])
```

```python
data_nan = data_sota_avg_test_all[['RAB_CS_DROP_RATE_3G','RAB_PS_DROP_RATE_3G','TBF_DROP_RATE_2G','TCH_DROP_RATE_2G']]
data_nan = data_nan.fillna(0)
array = np.array(data_nan)
array
```

```python
# Кластеризация сот
model = AgglomerativeClustering(n_clusters=3)
preds = model.fit_predict(array)
```

```python
preds
```

```python
data_nan['cluster'] = preds 
```
